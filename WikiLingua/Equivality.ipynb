```python
import pandas as pd
```

```python
text_grades = pd.read_csv("equivalency/equivalency_original.csv", index_col=False)
summary_grades = pd.read_csv("equivalency/equivalency_summary.csv", index_col=False)
```

```python
text_grades.head()
```

```python
summary_grades.head()
```

```python
columns = [column_name for column_name in summary_grades]
```

```python
summary_grades[columns[0]].tolist()
```

```python
from matplotlib import pyplot as plt
import matplotlib

matplotlib.rcParams.update({'font.size': 22})

fig, axs = plt.subplots(2, 2)
fig.set_figheight(15)
fig.set_figwidth(25)
fig.tight_layout()


for i, column_name in enumerate(columns):
    ind = (i // 2, i % 2)
    axs[*ind].set_title(column_name)
    axs[*ind].hist([abs(x - y) for x,y in zip(text_grades[column_name].tolist(), summary_grades[column_name].tolist())] , bins=[b - 0.5 for b in range(0, 6)], )
    axs[*ind].locator_params(axis='x', integer=True)
plt.show()
```
