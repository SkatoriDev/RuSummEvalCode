```python
summary_dataset = summary_dataset.rename(columns={"Unnamed: 0": "Ind"})
```

```python
index = "Ind"
title = "title"
article = "text"
ground_truth = "summary"

files = [
    f"{ground_truth}.txt", # Dummy for summary field
    "mbart_predictions.txt",
    "mt5_predictions.txt",
    "summarunner_predictions.txt",
]
summaries_fields = []
metrics = [
    "Актуальность",
    "Последовательность",
    "Беглость",
    "Согласованность",
]

for file in files:
    model_name = file.split(".")[0]
    summaries_fields.append(model_name)
    for metric in metrics:
        summaries_fields.append(f"{model_name}_{metric}")
    summaries_fields.append(f"{model_name}_Комментарий")

    
```

```python
summary_dataset = summary_dataset.reindex(columns = [index, article, *summaries_fields])
```

```python
summary_dataset.head()
```

```python
import random

random.seed(43)
indices_for_evaluation = random.sample(range(0, len(summary_dataset)), 55)
```

```python
indices_for_evaluation = sorted(indices_for_evaluation)
```

```python
print(indices_for_evaluation[:5])
```

```python
summary_dataset = summary_dataset[summary_dataset["Ind"].isin(indices_for_evaluation)]
```

## Join llm data

```python
summary_files = [
    "llama_7b_predictions.csv",
    "starling_predictions.csv",
    "yagpt_predictions.csv",
    "yagpt3_predictions.csv"
]

llm_data = pd.DataFrame()
for summary_file in summary_files:
    model_name = summary_file.split(".")[0]
    df = pd.read_csv(summary_file)[[model_name]]
    llm_data = pd.concat([llm_data, df], axis=1)

llm_data.head()
```

```python
summary_dataset.head()
print(len(llm_data))
```

```python
summary_dataset = pd.concat([summary_dataset, llm_data.set_index(summary_dataset.index)], axis=1)
```

```python
summary_dataset.head()
```

```python
index = "Ind"
title = "title"
article = "text"
ground_truth = "summary"

files = [
    f"{ground_truth}.txt", # Dummy for summary field
    "mbart_predictions.txt",
    "mt5_predictions.txt",
    "summarunner_predictions.txt",
] + summary_files
summaries_fields = []
metrics = [
    "Актуальность",
    "Последовательность",
    "Беглость",
    "Согласованность",
]

for file in files:
    model_name = file.split(".")[0]
    summaries_fields.append(model_name)
    for metric in metrics:
        summaries_fields.append(f"{model_name}_{metric}")
    summaries_fields.append(f"{model_name}_Комментарий")

    
```

```python
summary_dataset = summary_dataset.reindex(columns = [index, article, *summaries_fields])
```

```python
summary_dataset.head()
```

```python
summary_dataset.to_csv("summary_dataset_for_expert_evaluation_with_llm.csv", index=False)
```
