```python
import pandas as pd
```

```python
old_compiled = pd.read_csv("./human/old_compiled_gazeta.csv", index_col=False)
new_compiled = pd.read_csv("./human/new_compiled_gazeta.csv", index_col=False)
```

```python
len(old_compiled)
```

```python
new_compiled.head()
```

```python
old_compiled = old_compiled[old_compiled["Ind"].isin(list(new_compiled["Ind"]))]
```

```python
old_compiled.head()
```

```python
len(old_compiled)
```

```python
for i in range(len(old_compiled)):
    for field in new_compiled.columns:
        if field in old_compiled.columns and pd.isna(new_compiled.iloc[i][field]):
            new_compiled.at[i, field] = old_compiled.iloc[i][field]
print(new_compiled.columns)
new_compiled = new_compiled[[column for column in new_compiled.columns if "Комментарий" not in column]]
print(new_compiled.columns)
print(new_compiled.isna().any())
new_compiled = new_compiled.astype({field: int for field in new_compiled.columns if any(["Актуальность" in field, "Последовательность" in field, "Беглость" in field, "Согласованность" in field])})
```

```python
new_compiled.to_csv("total_compiled_gazeta.csv", index=False)
```
