```python
import pandas as pd
```

```python
gazeta_df = pd.read_csv("gazeta_merge_candidate.csv")
wikilingua_df = pd.read_csv("wikilingua_merge_candidate.csv")
```

```python
gazeta_df.head()
```

```python
wikilingua_df.head()
```

```python
merged_df = pd.concat([gazeta_df, wikilingua_df])
```

```python
merged_df.head()
```

```python
len(merged_df)
```

```python
bleu_df = merged_df[[column for column in merged_df.columns if "bleu" in column]]
rouge_df = merged_df[[column for column in merged_df.columns if "rouge1" in column]]
meteor_df = merged_df[[column for column in merged_df.columns if "meteor" in column]]
bertscore_df = merged_df[[column for column in merged_df.columns if "bertscore_f1" in column]]
```

```python
print(len(bleu_df), len(rouge_df), len(meteor_df), len(bertscore_df))
```

```python
bleu = {"bleu": list(bleu_df["mbart_predictions_bleu"]) + list(bleu_df["mt5_predictions_bleu"]) + list(bleu_df["summarunner_predictions_bleu"])}
rouge = {"rouge1": list(rouge_df["mbart_predictions_rouge1"]) + list(rouge_df["mt5_predictions_rouge1"]) + list(rouge_df["summarunner_predictions_rouge1"])}
meteor = {"meteor": list(meteor_df["mbart_predictions_meteor"]) + list(meteor_df["mt5_predictions_meteor"]) + list(meteor_df["summarunner_predictions_meteor"])}
bertscore = {"bertscore_f1": list(bertscore_df["mbart_predictions_bertscore_f1"]) + list(bertscore_df["mt5_predictions_bertscore_f1"]) + list(bertscore_df["summarunner_predictions_bertscore_f1"])}
```

```python
print(len(bleu), len(rouge), len(meteor), len(bertscore))
```

```python
metrics = pd.DataFrame({**bleu, **rouge, **meteor, **bertscore})
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib
%matplotlib inline

# load the Auto dataset

matplotlib.rcParams.update({'font.size': 30})
plt.subplots(figsize=(20,15))
# calculate the correlation matrix on the numeric columns
corr = metrics.select_dtypes('number').corr()

# plot the heatmap
sns.heatmap(corr, )
```
