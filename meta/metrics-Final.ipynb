```python
import pandas as pd
```

```python
data = pd.read_csv("metrics_data.csv")
```

```python
data.head()
```

```python
len(data)
```

```python
bleu_df = data[[column for column in data.columns if "bleu" in column]]
rouge_df = data[[column for column in data.columns if "rouge1" in column]]
meteor_df = data[[column for column in data.columns if "meteor" in column]]
bertscore_df = data[[column for column in data.columns if "bertscore_f1" in column]]
```

```python
print(len(bleu_df), len(rouge_df), len(meteor_df), len(bertscore_df))
```

```python
bleu = {"bleu": list(bleu_df["mbart_predictions_bleu"]) + list(bleu_df["mt5_predictions_bleu"]) + list(bleu_df["summarunner_predictions_bleu"]) + list(bleu_df["llama_7b_predictions_bleu"]) + list(bleu_df["starling_predictions_bleu"]) + list(bleu_df["yagpt_predictions_bleu"]) + list(bleu_df["yagpt3_predictions_bleu"])}
rouge = {"rouge1": list(rouge_df["mbart_predictions_rouge1"]) + list(rouge_df["mt5_predictions_rouge1"]) + list(rouge_df["summarunner_predictions_rouge1"]) + list(rouge_df["llama_7b_predictions_rouge1"]) + list(rouge_df["starling_predictions_rouge1"]) + list(rouge_df["yagpt_predictions_rouge1"]) + list(rouge_df["yagpt3_predictions_rouge1"])}
meteor = {"meteor": list(meteor_df["mbart_predictions_meteor"]) + list(meteor_df["mt5_predictions_meteor"]) + list(meteor_df["summarunner_predictions_meteor"]) + list(meteor_df["llama_7b_predictions_meteor"]) + list(meteor_df["starling_predictions_meteor"]) + list(meteor_df["yagpt_predictions_meteor"]) + list(meteor_df["yagpt3_predictions_meteor"])}
bertscore = {"bertscore_f1": list(bertscore_df["mbart_predictions_bertscore_f1"]) + list(bertscore_df["mt5_predictions_bertscore_f1"]) + list(bertscore_df["summarunner_predictions_bertscore_f1"])  + list(bertscore_df["llama_7b_predictions_bertscore_f1"]) + list(bertscore_df["starling_predictions_bertscore_f1"]) + list(bertscore_df["yagpt_predictions_bertscore_f1"]) + list(bertscore_df["yagpt3_predictions_bertscore_f1"])}
```

```python
print(len(bleu), len(rouge), len(meteor), len(bertscore))
```

```python
metrics = pd.DataFrame({**bleu, **rouge, **meteor, **bertscore})
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib
%matplotlib inline

# load the Auto dataset

matplotlib.rcParams.update({'font.size': 30})
plt.subplots(figsize=(20,15))
# calculate the correlation matrix on the numeric columns
corr = metrics.select_dtypes('number').corr()

# plot the heatmap
sns.heatmap(corr, )
```
